```python
import pandas as pd
import joblib
```

```python
# Load models
log_model = joblib.load("models/logistic_model.pkl")
xgb_model = joblib.load("models/xgboost_model.pkl")
lgb_model = joblib.load("models/lightgbm_model.pkl")

# Load feature columns (for tree models)
feature_cols = joblib.load("models/feature_columns.pkl")

print("✅ All models loaded successfully")
```

```python
test_input = {
    "purpose": "debtconsolidation",
    "isjointapplication": 0,
    "loanamount": 25000,
    "interestrate": 14.5,
    "monthlypayment": 590,
    "grade": "C2",
    "residentialstate": "CA",
    "homeownership": "rent",
    "annualincome": 52000,
    "incomeverified": 1,
    "dtiratio": 22.5,
    "lengthcredithistory": 8,
    "numtotalcreditlines": 14,
    "numopencreditlines": 10,
    "numopencreditlines1year": 4,
    "revolvingbalance": 15000,
    "revolvingutilizationrate": 72.0,
    "numderogatoryrec": 0,
    "numdelinquency2years": 1,
    "numchargeoff1year": 0,
    "numinquiries6mon": 2,
    "term_months": 60,
    "loan_year": 2016,
    "loan_month": 5,
    "loan_dayofweek": 2
}

test_df = pd.DataFrame([test_input])
```

```python
log_prob = log_model.predict_proba(test_df)[0][1]
log_pred = int(log_prob >= 0.5)
```

```python
# One-hot encode
test_enc = pd.get_dummies(test_df)

# Align with training features
test_enc = test_enc.reindex(columns=feature_cols, fill_value=0)
```

```python
xgb_prob = xgb_model.predict_proba(test_enc)[0][1]
xgb_pred = int(xgb_prob >= 0.5)
```

```python
lgb_prob = lgb_model.predict_proba(test_enc)[0][1]
lgb_pred = int(lgb_prob >= 0.5)
```

```python
results = pd.DataFrame({
    "Model": ["Logistic Regression", "XGBoost", "LightGBM"],
    "Default Probability": [log_prob, xgb_prob, lgb_prob],
    "Prediction (0=No Default, 1=Default)": [log_pred, xgb_pred, lgb_pred]
})

# print(results)
```
