```python
import pandas as pd

borrower = pd.read_csv("borrower.txt", sep="\t")
loan = pd.read_csv("loan.txt", sep="\t")
```

```python
borrower.to_csv("borrower.csv", index=False)
loan.to_csv("loan.csv", index=False)
```

```python
# run this first to see what the real columns are
print("Loan columns:", loan.columns.tolist())
print("Borrower columns:", borrower.columns.tolist())

# peek at the first few rows
display(loan.head())
display(borrower.head())
```

```python
loan_master = loan.merge(borrower, on="memberId", how="inner")
loan_master.to_csv("loan_master.csv", index=False)
```
